```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("final_data/basketball-final.sqlite")
```

```python
odds = pd.read_sql_query("SELECT * from BettingOdds_History", con)
```

```python
player_data = pd.read_csv("final_data/nba_games.csv")
```

```python
games = pd.read_sql_query("SELECT * from Game_FullDetails", con)
```

```python
games_playoffs = pd.read_sql_query("SELECT * from Game_Playoffs_FullDetails", con)
```

```python
total_games = pd.concat([games, games_playoffs], axis=0, ignore_index=True)
```

```python
total_games["GAME_ID"] = total_games["GAME_ID"].astype("object")
odds["GAME_ID"] = odds["GAME_ID"].astype("object")
```

```python
dataset = total_games.merge(odds, how="inner", on="GAME_ID")
```

```python
dataset["join_date"] = dataset["Date"].str.split(" ").str[0]
```

```python
dataset_player_dataset = pd.merge(dataset, player_data, how="inner", left_on=["join_date", "TEAM_ABBREVIATION_AWAY", "TEAM_ABBREVIATION_HOME"], right_on=["date", "team_opp", "team"])
```
